```python
semilla = int(input("Ingrese x0 con las condiciones son x0 igual o mayor que 0, A igual o mayor que cero, C igual o mayor que cero y M igual o mayor que X0"))
a = int(input("Ingrese A con las condiciones son x0 igual o mayor que 0, A igual o mayor que cero, C igual o mayor que cero y M igual o mayor que X0:"))
c =  int(input("Ingrese C con las condiciones son x0 igual o mayor que 0, A igual o mayor que cero, C igual o mayor que cero y M igual o mayor que X0"))
m = int(input("Ingrese m con las condiciones son x0 igual o mayor que 0, A igual o mayor que cero, C igual o mayor que cero y M igual o mayor que X0:"))

if semilla>=0 and a>= 0 and c>=0 and m>a:
  corridas = int (input("¿Cuantas corridas se va a generar?"))

  for x in range(corridas):
    semilla = semilla
    a = a
    c = c
    m = m
    sub = semilla * a + c
    pseudoal = sub % m
    random = pseudoal / (m-1)
    print(f"[{x}] x0:{semilla}, A:{a}, C:{c}, M:{m}, Sub:{sub}, Pseudoal:{pseudoal}, Random:{random}")
    semilla = pseudoal

    if pseudoal == 0:
      print("Limite del Metodo")
      break
    print ("Finalizado")

else:
  print("Error en los numeros introducidos, recuerde que las condiciones son x0 igual o mayor que 0, A igual o mayor que cero, C igual o mayor que cero y M igual o mayor que X0")
```
